```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('gnomad_benchmark.tsv', sep='\t', header=0)
```

```python
d
```

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.lineplot(data=d, x="N", y="queries_cost", hue="#tree_type", marker="o", ci=0).set(xscale="log")
plt.ylim(0, 5e8)
```

```python
sns.set(rc={'figure.figsize':(8,5)})
sns.lineplot(data=d, x="N", y="queries_ms", hue="#tree_type", marker="o", ci=0).set(xscale="log")
plt.ylim(0, 20000)
```
